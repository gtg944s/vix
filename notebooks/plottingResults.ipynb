```python
import pandas as pd
import numpy as np
from matplotlib.finance import candlestick_ohlc, candlestick2_ohlc
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('ggplot')
```

## Plot Results

```python
"""""""""

Function that plots OHLC chart 
input: 
    - df: pandas DataFrame with time, open, high, low, close as columns;
    - n_algo: number of algorithmic traders.
    
"""""""""
def ohlc_plot(df, n_algo):
    
    plt.figure(figsize=(16,9))
    
    ax1 = plt.subplot2grid((6,1), (0,0), rowspan=4, colspan=1, ylabel='price')
    ax2 = plt.subplot2grid((6,1), (4,0), rowspan=1, colspan=1, sharex=ax1,  xlabel='time')
    ax3 = plt.subplot2grid((6,1), (5,0), rowspan=1, colspan=1, sharex=ax1,  xlabel='time')
    
    ohlc = df[['time','open','high','low','close']].values
    candlestick_ohlc(ax1, ohlc, colorup = 'g')
    ax1.set_title('OHLC Chart - ' + str(n_algo) + '  Algo Traders', size=15)
    
    ax2.plot(df.volatilityIndex, color = 'coral')
    ax2.set_title('volatility index')
    
    ax3.plot(df.LogS, color = 'royalblue')
    ax3.set_title('Length of selling queue')
    plt.tight_layout()
```

- 50 Algorithmic Traders

```python
ohlc_plot(df_50, 50)
```

- 20 Algorithmic Traders

```python
ohlc_plot(df_20, 20)
```

- 5 Algorithmic Traders

```python
ohlc_plot(df_5, 5)
```
